```python
def findAllDefination(word):
    synonyms =[]
    for syn in wordnet.synsets(word):
        for l in syn.lemmas():
            synonyms.append(l.name())
    return synonyms
```

```python
word = "fun"
allDefination = findAllDefination(word)
print("All defination for word '"+  word+"' are" , set(allDefination))
```

```python
def findMeaning(word):
    #print(word)
    syns = wordnet.synsets(word)
    return syns[0].definition()
```

```python
for w in set(allDefination):
    allDef = findMeaning(w)
    print("Meaning of '"+ w+ "' is ",allDef)
```

```python
def displayAllDetails(word):
    syns = wordnet.synsets(word)
    data = []

    for s in syns:
        details = [s.name(),
                   ",".join([str(l.name()) for l in s.lemmas()]) ,
                   s.definition(),
                   s.examples(),
                   ",".join([str(l.name()) for l in s.hyponyms()]) ,
                   ",".join([str(l.name()) for l in s.hypernyms()]),
                   ",".join([str(l.name()) for l in s.lemmas()[0].antonyms()]) ,

                  ]
        data.append(details)
    df = pd.DataFrame(data, columns = ['Word with part of speech', 'Meanings', 'Defination','Examples','Hyponyms','Hypernyms','Antonyms'])

    return df

    
```

```python
data = displayAllDetails("play")
#print(data)
```

## Write  a program for feature extraction for above dataset with “Bag Of Word” method and  print the total words with their index .


```python
from sklearn.feature_extraction.text import CountVectorizer
  
document = ["Jim and Pam traveled by bus",
            "The train was late",
            "The flight was full. Traveling by flight is expensive"]
  
# Create a Vectorizer Object
vectorizer = CountVectorizer()
  
vectorizer.fit(document)
  
# Printing the identified Unique words along with their indices
print("Printing the identified Unique words along with their indices")
print("Vocabulary: ", vectorizer.vocabulary_)

print("\n Features",vectorizer.get_feature_names() )
  
# Encode the Document
vector = vectorizer.transform(document)
  
# Summarizing the Encoded Texts
print("\nEncoded Document is:")
print(vector.toarray())
```
